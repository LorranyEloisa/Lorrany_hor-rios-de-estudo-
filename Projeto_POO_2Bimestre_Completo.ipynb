## PARTE 2 ‚Äì M√©todos e Heran√ßa

**Tarefas:**

- Adicionar pelo menos 1 novo atributo e 1 novo m√©todo exclusivo na subclasse.
- Utilizar `super()` no construtor da subclasse.

**C√≥digo:**

```python

class Estudo:
    def __init__(self, disciplina, dia, hora):
        self.disciplina = disciplina
        self.dia = dia
        self.hora = hora

    def exibir_dados(self):
        print(f"Disciplina: {self.disciplina}")
        print(f"Dia: {self.dia}")
        print(f"Hora: {self.hora}")
        print("-" * 30)

    def mudar_hora(self, nova_hora):
        """Altera o hor√°rio do estudo"""
        self.hora = nova_hora
        print(f"A hora do estudo de {self.disciplina} foi alterada para {self.hora}")

    def mudar_dia(self, novo_dia):
        """Altera o dia do estudo"""
        self.dia = novo_dia
        print(f"O estudo de {self.disciplina} foi movido para {self.dia}")


class EstudoAvancado(Estudo):
    def __init__(self, disciplina, dia, hora, nivel):
        """Construtor da subclasse utilizando super()"""
        super().__init__(disciplina, dia, hora)
        self.nivel = nivel 
    def exibir_nivel(self):
        """Exibe o n√≠vel de estudo"""
        print(f"N√≠vel de estudo: {self.nivel}")

    def mudar_nivel(self, novo_nivel):
        """Altera o n√≠vel de estudo"""
        self.nivel = novo_nivel
        print(f"O n√≠vel de estudo foi alterado para: {self.nivel}")


estudo1 = Estudo("Matem√°tica", "segunda", "14:00")
estudo2 = Estudo("Hist√≥ria", "quarta", "16:30")
estudo_avancado1 = EstudoAvancado("F√≠sica", "sexta", "10:00", "Avan√ßado")

estudo1.exibir_dados()
estudo2.exibir_dados()

estudo1.mudar_hora("15:00")
estudo2.mudar_dia("sexta")
estudo1.exibir_dados()
estudo2.exibir_dados()

estudo_avancado1.exibir_dados()
estudo_avancado1.exibir_nivel()
estudo_avancado1.mudar_nivel("Intermedi√°rio")
estudo_avancado1.exibir_nivel()
```

```python
# Importando o m√≥dulo sqlite3 para integra√ß√£o com o banco de dados SQLite
import sqlite3

# -----------------------------------------------
# PARTE 1 - Classe Principal e Objetos
# -----------------------------------------------

class Estudo:
    def __init__(self, disciplina, dia, hora):
        """
        M√©todo construtor que inicializa os atributos de cada objeto.
        :param disciplina: Nome da disciplina de estudo.
        :param dia: Dia da semana em que o estudo ocorre.
        :param hora: Hora em que o estudo √© programado.
        """
        self.disciplina = disciplina
        self.dia = dia
        self.__hora = hora  # Atributo privado: hora do estudo.

    def exibir_dados(self):
        """Exibe os dados do estudo formatados."""
        print(f"Disciplina: {self.disciplina}")
        print(f"Dia: {self.dia}")
        print(f"Hora: {self.__hora}")
        print("-" * 30)

    def set_hora(self, nova_hora):
        """
        M√©todo set para alterar a hora de estudo, com valida√ß√£o de formato.
        :param nova_hora: A nova hora a ser atribu√≠da ao atributo __hora.
        """
        if self.validar_hora(nova_hora):
            self.__hora = nova_hora
            print(f"A hora foi alterada para {self.__hora}")
        else:
            print("Hora inv√°lida. Hora n√£o alterada.")

    def validar_hora(self, hora):
        """
        Valida se a hora est√° no formato correto (HH:MM).
        :param hora: Hora a ser validada.
        :return: True se for v√°lida, False caso contr√°rio.
        """
        try:
            h, m = map(int, hora.split(":"))
            return 0 <= h < 24 and 0 <= m < 60
        except ValueError:
            return False


# -----------------------------------------------
# PARTE 2 - M√©todos e Heran√ßa
# -----------------------------------------------

class EstudoAvancado(Estudo):
    def __init__(self, disciplina, dia, hora, nivel):
        """
        M√©todo construtor da subclasse EstudoAvancado.
        :param disciplina: Nome da disciplina de estudo.
        :param dia: Dia da semana.
        :param hora: Hora do estudo.
        :param nivel: N√≠vel de dificuldade do estudo (ex: b√°sico, intermedi√°rio, avan√ßado).
        """
        super().__init__(disciplina, dia, hora)
        self.nivel = nivel  # Atributo exclusivo da subclasse EstudoAvancado.

    def exibir_dados(self):
        """Sobrescreve o m√©todo exibir_dados para incluir o n√≠vel de estudo."""
        print(f"Disciplina (Avan√ßado): {self.disciplina}")
        print(f"Dia (Avan√ßado): {self.dia}")
        print(f"Hora (Avan√ßado): {self._Estudo__hora}")
        print(f"N√≠vel de Estudo: {self.nivel}")
        print("-" * 30)

    def mudar_nivel(self, novo_nivel):
        """M√©todo exclusivo para alterar o n√≠vel de estudo."""
        self.nivel = novo_nivel
        print(f"O n√≠vel de estudo foi alterado para: {self.nivel}")


# -----------------------------------------------
# PARTE 3 - Encapsulamento e Abstra√ß√£o
# -----------------------------------------------

# Aqui, o atributo __hora foi tornado privado na classe Estudo
# M√©todos get/set s√£o criados para interagir com o atributo privado.

# -----------------------------------------------
# PARTE 4 - Polimorfismo e Banco de Dados
# -----------------------------------------------

# Fun√ß√µes de banco de dados

def criar_conexao_db():
    """Cria e retorna uma conex√£o com o banco de dados SQLite."""
    conn = sqlite3.connect('agenda_estudos.db')
    return conn

def criar_tabela(conn):
    """Cria a tabela de estudos no banco de dados."""
    cursor = conn.cursor()
    cursor.execute('''
    CREATE TABLE IF NOT EXISTS estudos (
        id INTEGER PRIMARY KEY,
        disciplina TEXT,
        dia TEXT,
        hora TEXT
    )
    ''')
    conn.commit()

def inserir_estudo(conn, disciplina, dia, hora):
    """Insere um novo registro de estudo na tabela."""
    cursor = conn.cursor()
    cursor.execute('''
    INSERT INTO estudos (disciplina, dia, hora)
    VALUES (?, ?, ?)
    ''', (disciplina, dia, hora))
    conn.commit()

def consultar_estudos(conn):
    """Consulta todos os registros de estudos no banco de dados."""
    cursor = conn.cursor()
    cursor.execute('SELECT * FROM estudos')
    return cursor.fetchall()

def excluir_estudo(conn, estudo_id):
    """Exclui um registro de estudo pela ID."""
    cursor = conn.cursor()
    cursor.execute('DELETE FROM estudos WHERE id = ?', (estudo_id,))
    conn.commit()


# -----------------------------------------------
# PARTE 5 - Projeto Final e Organiza√ß√£o
# -----------------------------------------------

def main():
    # Criando objetos da classe Estudo
    estudo1 = Estudo("Matem√°tica", "segunda", "14:00")
    estudo2 = Estudo("Hist√≥ria", "quarta", "16:30")

    # Exibindo dados dos estudos
    estudo1.exibir_dados()
    estudo2.exibir_dados()

    # Alterando hora e dia dos estudos
    estudo1.set_hora("15:00")
    estudo2.set_hora("17:00")
    estudo1.set_hora("25:00")  # Tentando hora inv√°lida

    # Criando um objeto da subclasse EstudoAvancado
    estudo_avancado1 = EstudoAvancado("F√≠sica", "sexta", "10:00", "Avan√ßado")
    estudo_avancado1.exibir_dados()

    # Alterando o n√≠vel de estudo na subclasse
    estudo_avancado1.mudar_nivel("Intermedi√°rio")
    estudo_avancado1.exibir_dados()

    # Integra√ß√£o com banco de dados
    conn = criar_conexao_db()
    criar_tabela(conn)

    # Inserindo estudos no banco de dados
    inserir_estudo(conn, "Matem√°tica", "segunda", "14:00")
    inserir_estudo(conn, "Hist√≥ria", "quarta", "16:30")

    # Consultando todos os registros de estudo
    estudos = consultar_estudos(conn)
    print("Estudos no banco de dados:")
    for estudo in estudos:
        print(estudo)

    # Excluindo um estudo pelo ID
    excluir_estudo(conn, 1)

    # Consultando novamente ap√≥s exclus√£o
    estudos = consultar_estudos(conn)
    print("Estudos ap√≥s exclus√£o:")
    for estudo in estudos:
        print(estudo)

    # Fechando a conex√£o com o banco de dados
    conn.close()

if __name__ == "__main__":
    main()
```
